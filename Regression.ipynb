```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_simple_dataset(n_samples = 100, n_features = 1, noise = 20, show_plot=True):
    X, y = datasets.make_regression(n_samples=n_samples, n_features=n_features, n_informative=1, noise=noise, random_state=37)
    
    if show_plot:
        fig = plt.figure(figsize=(4, 4))
        plt.xlabel("x (feature)")
        plt.ylabel("y (output)")
        plt.title("Synthetic data set")
        plt.scatter(X, y)
        plt.show()
    
    return X, y

def generate_complex_dataset(n_samples = 100):
    x = np.linspace(-1, 1, n_samples) # coordinates
    noise_sample = np.random.normal(0,0.5,n_samples)
    sine_wave = x + np.sin(4*x) + noise_sample
    plt.plot(x, sine_wave, 'o');
    plt.show()
    
    return x, sine_wave
```

### 1.1 Closed-Form

```python
class LinearRegression(object):
    def fit(self, X, t):
        N, D = X.shape
        
        self.w = np.linalg.pinv(X) @ t
        
        self.var = np.mean(np.square(X @ self.w - t))

    def predict(self, X, return_std=False):
        N, D = X.shape
        
        y = X @ self.w
        
        if return_std:
            y_std = np.sqrt(self.var)
            
            return y, y_std
        return y
```

```python
def extract_polynomial_features(X, M):
    phi = []
    for x in X:
        # phi_n = [1, x, x^2, ..., x^M]
        phi.append(x ** np.arange(M + 1))

    # phi - np.array de dimensiune (N, M+1), unde N este numarul de linii din X
    phi = list(map(lambda x: x ** range(M + 1), X))
    return np.array(phi)
```

### 1.2 Gradient Descent

```python
class GradientDescent(object):
    def fit(self, X, t, lr=1.00e-3, epochs_no=50, print_epochs=True):
        N, D = X.shape
        
        vs = []
        
        self.w = np.random.randn((D))
        for i in range(epochs_no):
            # y = np.dot(X, self.w)
            # self.w -= (lr * np.dot(y - t, X))
            
            y = X @ self.w
            self.w -= (lr * (X.T @ (y - t)))
        
            self.var = np.mean(np.square(X @ self.w - t))
            vs.append(self.var)
            # print(self.var)
        
        if (print_epochs):
            plt.xlabel("number of iterations")
            plt.ylabel("empirical risk")
            plt.title("Batch Gradient Descent")
            plt.scatter(list(range(50)), vs)
            # plt.plot(list(range(50)), vs, "blue", label="data points")
            plt.plot(list(range(50)), vs, label="lr - " + str(lr))
            plt.legend()

        print(self.w)

    def predict(self, X, return_std=False):
        N, D = X.shape
        
        y = X @ self.w
        
        if return_std:
            y_std = np.sqrt(self.var)
            
            return y, y_std
        return y
```

### 1.3 Mini-Batch Gradient Descent

```python
class MiniBatchGradientDescent(object):
    def fit(self, X, t, M, lr=1.00e-3, epochs_no=50, print_epochs=False):
        N, D = X.shape
        
        batch = 0
        vs = []
        
        self.w = np.random.randn((D))
        for i in range(epochs_no):
            if ((batch + M) >= N):
                batch = 0
            
            X_batch = X[batch:(batch + M)]
            t_batch = t[batch:(batch + M)]
            
            batch += 1
            
            # y = np.dot(X, self.w)
            # self.w -= (lr * np.dot(y - t, X))
            y = X_batch @ self.w
            
            # self.w -= (lr * ((1 / M) * (X_batch.T @ (y - t_batch))))
            self.w -= (lr * ((X_batch.T @ (y - t_batch))))
        
            self.var = np.mean(np.square(X @ self.w - t))
            vs.append(self.var)
        
        if (print_epochs):
            plt.xlabel("number of iterations")
            plt.ylabel("empirical risk")
            plt.title("Mini-Batch Gradient Descent - M=" + str(M))
            plt.scatter(list(range(50)), vs)
            plt.plot(list(range(50)), vs, label="lr - " + str(lr))
            plt.legend()

        # print(self.w)

    def predict(self, X, return_std=False):
        N, D = X.shape
        
        y = X @ self.w
        
        if return_std:
            y_std = np.sqrt(self.var)
            
            return y, y_std
        return y
```

### 1.4 Stochastic Gradient Descent

```python
from sklearn.utils import shuffle

class StochasticGradientDescent(object):
    def fit(self, X, t, lr=1.00e-3, epochs_no=50, print_epochs=True):
        N, D = X.shape
        
        s_X = np.copy(X)
        s_t = np.copy(t)
        
        vs = []
        
        self.w = np.random.randn((D))
        for i in range(epochs_no):
            # "shuffle" the data
            s_X, s_t = shuffle(s_X, s_t, random_state = 37)
             
            X_batch = s_X[0]
            t_batch = s_t[0]
            
            # y = np.dot(X, self.w)
            # self.w -= (lr * np.dot(y - t, X))
            y = X_batch @ self.w
            
            self.w -= (lr * ((X_batch.T * (y - t_batch))))
        
            self.var = np.mean(np.square(X @ self.w - t))
            vs.append(self.var)
        
        if (print_epochs):
            plt.xlabel("number of iterations")
            plt.ylabel("empirical risk")
            plt.title("Stochastic Gradient Descent")
            plt.scatter(list(range(50)), vs)
            plt.plot(list(range(50)), vs, label="lr - " + str(lr))
            plt.legend()

        print(self.w)

    def predict(self, X, return_std=False):
        N, D = X.shape
        
        y = X @ self.w
        
        if return_std:
            y_std = np.sqrt(self.var)
            
            return y, y_std
        return y
```

## 2 - K-Fold Validation; Ridge Regression

```python
class RidgeRegression(object):    
    def fit(self, X, t, alpha):
        N, D = X.shape
        self.w = np.zeros(D)
        
        I = np.eye(D)
        a = alpha * I
        
        self.w = np.linalg.inv(X.T @ X + a) @ X.T @ t
        
        self.var = np.mean(np.square(X @ self.w - t))
    
    def predict(self, X, return_std=False):
        N, D = X.shape
        
        y = X @ self.w
        
        if return_std:
            y_std = np.sqrt(self.var)
            
            return y, y_std
        return y
```

```python
class KFoldValidation(object):
    def evaluate(self, model, X, t, alpha=0, tests_no=100, k=10, show_plot=True):
        N, D = X.shape
        step = N // k
        
        s_X = np.copy(X)
        s_t = np.copy(t)
        
        err = 0
        es = []
        
        for _ in range(tests_no):
            s_X, s_t = shuffle(s_X, s_t, random_state=0)
            test_err = 0
            
            for i in range(k):
                l = (i * step)
                r = (i + 1) * step
                
                X_test = s_X[l:r]
                t_test = s_t[l:r]
                
                X_train = np.concatenate([s_X[0:l], s_X[r:]], axis=0)
                t_train = np.concatenate([s_t[0:l], s_t[r:]], axis=0)
                model.fit(X_train, t_train, alpha)
                
                y_test = model.predict(X_test)
                ey = mean_squared_error(y_test, t_test)
                # print("Test error:", ey)
                
                test_err += ey
                err += ey
            
            test_err /= k
            es.append(test_err)
                    
        if (show_plot):
            plt.xlabel("number of iterations")
            plt.ylabel("error")
            plt.title("K-Fold Validation - k=" + str(k))
            plt.scatter(list(range(tests_no)), es)
            plt.plot(list(range(tests_no)), es, label="alpha=" + str(alpha))
            plt.legend(bbox_to_anchor=(1, 1), loc='upper right', ncol=1)
            # plt.legend()
            
        err /= (tests_no * k)
        # print("Mean error:", err)
        return err
```

```python
import csv
YVAR = "_y"

def parse_csv_dataset(fname):
    fin = open(fname, 'r')
    csv_reader = csv.reader(fin, delimiter=";")

    row = next(csv_reader)
    attribute_names = row[:(-1)]
    
    examples = []
    ys = set()

    for row in csv_reader:
        *attributes, target = row
        
        attributes = [float(x) for x in attributes]
        target = float(target)
        
        ys.add(target)

        example = dict(zip(attribute_names, attributes))
        example[YVAR] = target
        examples.append(example)

    fin.close()
    return ys, attribute_names, examples
```

```python
class RegressionNode:
    def __init__(self, label, threshold):
        self.label = label
        self.threshold = threshold
        
        # self.left = None
        # self.right = None
        self.children = {}
    
    def display(self, indent = ""):
        print(indent + (self.label + ":" if (self.children) else "<" + str(self.label) + ">"))
        indent += "   "
        if self.children:
            for key, value in self.children.items():
                print(indent + ":" + str(key))
                value.display(indent + "   ")
```

```python
import math

def retrieve_attribute_cut_values(X, a):
    sc = set()
    
    for x in X:
        sc.add(x[a])
    
    # convert to list
    lc = sorted(sc)
    
    cuts = [((b + c) / 2) for b, c in zip(lc, lc[1:])]
    if (len(cuts) == 0):
        # only one value for _a_ in X -- |lc| == 1
        cuts = lc
    
    return cuts

def calculate_rss_attribute_cut(X, a, c):
    y_left = [x[YVAR] for x in X if (x[a] <= c)]
    y_right = [x[YVAR] for x in X if (x[a] > c)]
    
    if (len(y_left) > 0):
        y_left_mean = np.mean(y_left)
    
    if (len(y_right) > 0):
        y_right_mean = np.mean(y_right)
    
    rss = 0.0
    for y in y_left:
        aux = (y - y_left_mean) ** 2
        rss += aux
    
    for y in y_right:
        aux = (y - y_right_mean) ** 2
        rss += aux
    
    return rss

def calculate_min_rss_attribute(X, a):
    cuts = retrieve_attribute_cut_values(X, a)
    
    min_rss = math.inf
    min_cut = None
    
    for c in cuts:
        rss_c = calculate_rss_attribute_cut(X, a, c)
        
        if (rss_c < min_rss):
            min_rss = rss_c
            min_cut = c
    
    return min_rss, min_cut

def build_regression_tree(X, A):
    # check leaf (1) - |X| <= 5
    if (len(X) <= 5):
        y_mean = 0.0
        
        for x in X:
            y_mean += x[YVAR]
        
        y_mean /= (len(X))
        return RegressionNode(y_mean, y_mean)
    
    min_rss = math.inf
    min_a = None
    min_cut = None
    
    for a in A:
        rss_a, cut_a = calculate_min_rss_attribute(X, a)
        
        if (rss_a < min_rss):
            min_rss = rss_a
            min_a = a
            min_cut = cut_a
    
    node_a = RegressionNode(min_a, min_cut)
    
    # check leaf (2) - min_rss == 0
    if (min_rss == 0):
        yl = [x[YVAR] for x in X if (x[min_a] <= min_cut)]
        yr = [x[YVAR] for x in X if (x[min_a] > min_cut)]
        
        if (yl[0] == yr[0]):
            node_a = RegressionNode(yl[0], yl[0])
            return node_a
        
        if (len(yl) > 0):
            y_mean_l = yl[0]
            node_a.children["left"] = RegressionNode(y_mean_l, y_mean_l)
        
        if (len(yr) > 0):
            y_mean_r = yr[0]
            node_a.children["right"] = RegressionNode(y_mean_r, y_mean_r)
        
        return node_a
    
    Xl = [x for x in X if (x[min_a] <= min_cut)]
    Xr = [x for x in X if (x[min_a] > min_cut)]
    
    node_a.children["left"] = build_regression_tree(Xl, A)
    node_a.children["right"] = build_regression_tree(Xr, A)
    
    return node_a
```

```python
def calculate_node_mean(X, node):
    y_mean = 0.0
        
    for x in X:
        y_mean += x[YVAR]

    y_mean /= (len(X))
    return y_mean

def calculate_node_rss(X, node):
    y_mean = calculate_node_mean(X, node)
    
    rss = 0.0
    for x in X:
        aux = (x[YVAR] - y_mean) ** 2
        rss += aux
    
    return rss

def find_best_pruning_node(X, node):
    if (len(node.children) > 0):
        node_rss = calculate_node_rss(X, node)

        Xl = [x for x in X if (x[node.label] <= node.threshold)]
        Xr = [x for x in X if (x[node.label] > node.threshold)]

        g_cl, min_cl, tree_rss_l, node_leaves_l = find_best_pruning_node(Xl, node.children["left"])
        g_cr, min_cr, tree_rss_r, node_leaves_r = find_best_pruning_node(Xr, node.children["right"])
        
        tree_rss = tree_rss_l + tree_rss_r
        node_leaves = node_leaves_l + node_leaves_r
        
        g_node = (node_rss - tree_rss) / (node_leaves - 1)

        min_g = g_node
        min_node = node

        if (g_cl < min_g):
            min_g = g_cl
            min_node = min_cl

        if (g_cr < min_g):
            min_g = g_cr
            min_node = min_cr

        return min_g, min_node, tree_rss, node_leaves
    
    # node -- leaf
    rss = 0.0
    for x in X:
        aux = (x[YVAR] - node.label) ** 2
        rss += aux
    
    # don't want to prune a leaf (!) --> math.inf
    return math.inf, None, rss, 1

def prune_node_subtree(X, root, node):
    if (root == node):
        # do the pruning
        y_mean = calculate_node_mean(X, node)
        root_copy = RegressionNode(y_mean, y_mean)
        
        return root_copy
    
    root_copy = RegressionNode(root.label, root.threshold)
    
    if (len(root.children) == 0):
        return root_copy
    
    Xl = [x for x in X if (x[root.label] <= root.threshold)]
    Xr = [x for x in X if (x[root.label] > root.threshold)]
    
    root_copy.children["left"] = prune_node_subtree(Xl, root.children["left"], node)
    root_copy.children["right"] = prune_node_subtree(Xr, root.children["right"], node)
    
    return root_copy

def retrieve_pruning_alpha_trees(X, root):
    if (len(root.children) == 0):
        return [], []
    
    alpha, node, _, _ = find_best_pruning_node(X, root)
    tree = prune_node_subtree(X, root, node)
    
    trees, alphas = retrieve_pruning_alpha_trees(X, tree)
    
    trees.append(tree)
    alphas.append(alpha)
    
    return trees, alphas
```

```python
def find_alpha_pruning_node(X, node, alpha):
    if (len(node.children) > 0):
        node_rss = calculate_node_rss(X, node)

        Xl = [x for x in X if (x[node.label] <= node.threshold)]
        Xr = [x for x in X if (x[node.label] > node.threshold)]

        cost_cl, min_cl, tree_rss_l, node_leaves_l = find_alpha_pruning_node(Xl, node.children["left"], alpha)
        cost_cr, min_cr, tree_rss_r, node_leaves_r = find_alpha_pruning_node(Xr, node.children["right"], alpha)
        
        tree_rss = tree_rss_l + tree_rss_r
        node_leaves = node_leaves_l + node_leaves_r
        
        # the total cost by pruning would be
        # RSS_all + (node_rss - tree_rss) + alpha * (all_leaves - node_leaves + 1)
        # can remove the constants (!)
        cost_node = (node_rss - tree_rss) - alpha * node_leaves

        min_cost = cost_node
        min_node = node

        if (cost_cl < min_cost):
            min_cost = cost_cl
            min_node = min_cl

        if (cost_cr < min_cost):
            min_cost = cost_cr
            min_node = min_cr

        return min_cost, min_node, tree_rss, node_leaves
    
    # node -- leaf
    rss = 0.0
    for x in X:
        aux = (x[YVAR] - node.label) ** 2
        rss += aux
    
    # don't want to prune a leaf (!) --> math.inf
    return math.inf, None, rss, 1

def evaluate_rss(X, root):
    if (len(root.children) == 0):
        rss = 0.0
        
        for x in X:
            aux = (x[YVAR] - root.label) ** 2
            rss += aux
        
        return rss
    
    Xl = [x for x in X if (x[root.label] <= root.threshold)]
    Xr = [x for x in X if (x[root.label] > root.threshold)]
    
    rss_l = evaluate_rss(Xl, root.children["left"])
    rss_r = evaluate_rss(Xr, root.children["right"])
    
    return rss_l + rss_r

def k_fold_cross_validation(X, K = 10):
    flen = math.ceil(len(X) / K)
    
    a_err = [0 for _ in alphas]
    
    for k in range(K - 1):
        l = k * flen
        r = (k + 1) * flen
        
#         print(l, " - ", r)
        
        X_train = np.concatenate([X[:l], X[r:]], axis = 0)
        X_test = X[l:r]
        
        k_root = build_regression_tree(X_train, wine_attribute_names)
        
        a_k_err = []
        
        # alpha = 0 --> k_root
        a_rss = evaluate_rss(X_test, k_root)
        a_k_err.append(a_rss)
        
        for alpha in alphas[1:]:
            a_cost, a_node, a_t_rss, a_lvs = find_alpha_pruning_node(X_train, k_root, alpha)
            a_tree = prune_node_subtree(X_train, k_root, a_node)
            
            a_rss = evaluate_rss(X_test, a_tree)
            a_rss /= len(X_test)
            
            a_k_err.append(math.sqrt(a_rss))
            
#             if (k == 0):
#                 print(a_node.label)
        
        for i in range(len(alphas)):
            a_err[i] += a_k_err[i]

    k = K - 1
    l = k * flen

    X_train = X[:l]
    X_test = X[l:]

    k_root = build_regression_tree(X_train, wine_attribute_names)

    a_k_err = []
    
    # alpha = 0 --> k_root
    a_rss = evaluate_rss(X_test, k_root)
    a_k_err.append(a_rss)

    for alpha in alphas[1:]:
        a_cost, a_node, _, _ = find_alpha_pruning_node(X_train, k_root, alpha)
        a_tree = prune_node_subtree(X_train, k_root, a_node)

        a_rss = evaluate_rss(X_test, a_tree)
        a_rss /= len(X_test)
        a_k_err.append(math.sqrt(a_rss))

    for i in range(len(alphas)):
        a_err[i] += a_k_err[i]
    
    min_err = math.inf
    min_pos = None

    pos = 0
    for err in a_err:
        if (err < min_err):
            min_err = err
            min_pos = pos
        
        pos += 1
    
    print(min_err, min_pos)
    return min_err, min_pos
```

```python
import re

def parse_spaced_dataset(fname, attribute_names):
    with open(fname) as file:
        lines = [re.split("\t|\s\s+", line.rstrip())[:-1] for line in file]
    
    examples = []
    ys = set()
    
    for row in lines:
        target, *attributes = row
        
        # remove the rows with missing values
        if (attributes[2] == "?"):
            continue
        
        attributes = [float(x) for x in attributes]
        target = float(target)
        
        ys.add(target)

        example = dict(zip(attribute_names, attributes))
        example[YVAR] = target
        examples.append(example)
    
    return ys, examples
```

```python
# Calculate the RSS of a discrete attribute
# ^ separate into all of its discrete values
def calculate_rss_attribute_discrete(X, a):
    vals = set()
    
    for x in X:
        vals.add(x[a])
    
    rss = 0.0
    for v in vals:
        yv = [x[YVAR] for x in X if (x[a] == v)]
        
        yv_mean = np.mean(yv)
        for y in yv:
            aux = (y - yv_mean) ** 2
            rss += aux
    
    return rss

def build_multi_regression_tree(X, A, discrete_A):
    # check leaf (1) - |X| <= 5
    if (len(X) <= 5):
        y_mean = 0.0
        
        for x in X:
            y_mean += x[YVAR]
        
        y_mean /= (len(X))
        return RegressionNode(y_mean, y_mean)
    
    min_rss = math.inf
    min_a = None
    min_cut = None
    
    for a in A:
        if (a in discrete_A):
            rss_a = calculate_rss_attribute_discrete(X, a)
            cut_a = None
        else:
            rss_a, cut_a = calculate_min_rss_attribute(X, a)

        if (rss_a < min_rss):
            min_rss = rss_a
            min_a = a
            min_cut = cut_a
    
    node_a = RegressionNode(min_a, min_cut)
    
    # check leaf (2) - min_rss == 0
    if (min_rss == 0):
        y = [x[YVAR] for x in X]
        y_mean = np.mean(y)
        
        return RegressionNode(y_mean, y_mean)
    
    if (min_a in discrete_A):
        vals = set()
    
        for x in X:
            vals.add(x[min_a])
            
        for v in vals:
            Xv = [x for x in X if (x[min_a] == v)]
            node_a.children[v] = build_multi_regression_tree(Xv, A, discrete_A)
        
    else:
        Xl = [x for x in X if (x[min_a] <= min_cut)]
        Xr = [x for x in X if (x[min_a] > min_cut)]

        node_a.children["left"] = build_multi_regression_tree(Xl, A, discrete_A)
        node_a.children["right"] = build_multi_regression_tree(Xr, A, discrete_A)
    
    return node_a
```

```python
def find_best_multi_pruning_node(X, node):
    if (len(node.children) > 0):
        node_rss = calculate_node_rss(X, node)
        
        tree_rss = 0.0
        node_leaves = 0
        
        min_g = math.inf
        min_node = None
        
        # check if node is discrete or continuous
        if (node.threshold is None):
            vals = set()
    
            for x in X:
                vals.add(x[node.label])

            for v in vals:
                Xv = [x for x in X if (x[node.label] == v)]
                g_c, min_c, tree_rss_c, node_leaves_c = find_best_multi_pruning_node(Xv, node.children[v])
                
                if (g_c < min_g):
                    min_g = g_c
                    min_node = min_c
                
                tree_rss += tree_rss_c
                node_leaves += node_leaves_c
        else:
            Xl = [x for x in X if (x[node.label] <= node.threshold)]
            Xr = [x for x in X if (x[node.label] > node.threshold)]

            g_cl, min_cl, tree_rss_l, node_leaves_l = find_best_multi_pruning_node(Xl, node.children["left"])
            g_cr, min_cr, tree_rss_r, node_leaves_r = find_best_multi_pruning_node(Xr, node.children["right"])
            
            if (g_cl < min_g):
                min_g = g_cl
                min_node = min_cl

            if (g_cr < min_g):
                min_g = g_cr
                min_node = min_cr

            tree_rss = tree_rss_l + tree_rss_r
            node_leaves = node_leaves_l + node_leaves_r
        
        g_node = (node_rss - tree_rss) / (node_leaves - 1)

        if (g_node < min_g):
            min_g = g_node
            min_node = node

        return min_g, min_node, tree_rss, node_leaves
    
    # node -- leaf
    rss = 0.0
    for x in X:
        aux = (x[YVAR] - node.label) ** 2
        rss += aux
    
    # don't want to prune a leaf (!) --> math.inf
    return math.inf, None, rss, 1

def prune_multi_node_subtree(X, root, node):
    if (root == node):
        # do the pruning
        y_mean = calculate_node_mean(X, node)
        root_copy = RegressionNode(y_mean, y_mean)
        
        return root_copy
    
    root_copy = RegressionNode(root.label, root.threshold)
    if (len(root.children) == 0):
        return root_copy
    
    if (root.threshold is None):
        vals = set()
    
        for x in X:
            vals.add(x[root.label])

        for v in vals:
            Xv = [x for x in X if (x[root.label] == v)]
            root_copy.children[v] = prune_multi_node_subtree(Xv, root.children[v], node)
            
    else:
        Xl = [x for x in X if (x[root.label] <= root.threshold)]
        Xr = [x for x in X if (x[root.label] > root.threshold)]

        root_copy.children["left"] = prune_multi_node_subtree(Xl, root.children["left"], node)
        root_copy.children["right"] = prune_multi_node_subtree(Xr, root.children["right"], node)
    
    return root_copy

def retrieve_multi_pruning_alpha_trees(X, root):
    if (len(root.children) == 0):
        return [], []
    
    alpha, node, _, _ = find_best_multi_pruning_node(X, root)
    tree = prune_multi_node_subtree(X, root, node)
    
    trees, alphas = retrieve_multi_pruning_alpha_trees(X, tree)
    
    trees.append(tree)
    alphas.append(alpha)
    
    return trees, alphas
```

```python
def evaluate_multi_rss(X, root):
    if (len(root.children) == 0):
        rss = 0.0
        
        for x in X:
            aux = (x[YVAR] - root.label) ** 2
            rss += aux
        
        return rss
    
    rss = 0.0
    
    if (root.threshold is None):
        vals = set()
    
        for x in X:
            vals.add(x[root.label])

        for v in vals:
            Xv = [x for x in X if (x[root.label] == v)]
            
            rss_c = 0.0
            if (v in root.children):
                rss_c = evaluate_multi_rss(Xv, root.children[v])
            else:
                for x_v in Xv:
                    rss_c += (x_v[YVAR] ** 2)
            
            rss += rss_c
    else:
        Xl = [x for x in X if (x[root.label] <= root.threshold)]
        Xr = [x for x in X if (x[root.label] > root.threshold)]

        rss_l = evaluate_multi_rss(Xl, root.children["left"])
        rss_r = evaluate_multi_rss(Xr, root.children["right"])
        
        rss = rss_l + rss_r
    
    return rss
```

```python
def find_alpha_multi_pruning_node(X, node, alpha):
    if (len(node.children) > 0):
        node_rss = calculate_node_rss(X, node)
        
        tree_rss = 0.0
        node_leaves = 0
        
        min_cost = math.inf
        min_node = None
        
        if (node.threshold is None):
            vals = set()
    
            for x in X:
                vals.add(x[node.label])

            for v in vals:
                Xv = [x for x in X if (x[node.label] == v)]
                cost_c, min_c, tree_rss_c, node_leaves_c = find_alpha_multi_pruning_node(Xv, node.children[v], alpha)
                
                if (cost_c < min_cost):
                    min_cost = cost_c
                    min_node = min_c
                
                tree_rss += tree_rss_c
                node_leaves += node_leaves_c
        else:
            Xl = [x for x in X if (x[node.label] <= node.threshold)]
            Xr = [x for x in X if (x[node.label] > node.threshold)]

            cost_cl, min_cl, tree_rss_l, node_leaves_l = find_alpha_multi_pruning_node(Xl, node.children["left"], alpha)
            cost_cr, min_cr, tree_rss_r, node_leaves_r = find_alpha_multi_pruning_node(Xr, node.children["right"], alpha)
        
            tree_rss = tree_rss_l + tree_rss_r
            node_leaves = node_leaves_l + node_leaves_r
            
            if (cost_cl < min_cost):
                min_cost = cost_cl
                min_node = min_cl

            if (cost_cr < min_cost):
                min_cost = cost_cr
                min_node = min_cr
        
        # the total cost by pruning would be
        # RSS_all + (node_rss - tree_rss) + alpha * (all_leaves - node_leaves + 1)
        # can remove the constants (!)
        cost_node = (node_rss - tree_rss) - alpha * node_leaves
        
        if (cost_node < min_cost):
            min_cost = cost_node
            min_node = node

        return min_cost, min_node, tree_rss, node_leaves
    
    # node -- leaf
    rss = 0.0
    for x in X:
        aux = (x[YVAR] - node.label) ** 2
        rss += aux
    
    # don't want to prune a leaf (!) --> math.inf
    return math.inf, None, rss, 1

def k_fold_multi_cross_validation(X, K = 10):
    flen = math.ceil(len(X) / K)
    
    a_err = [0 for _ in auto_alphas]
    
    for k in range(K - 1):
        l = k * flen
        r = (k + 1) * flen
        
        X_train = np.concatenate([X[:l], X[r:]], axis = 0)
        X_test = X[l:r]
        
        k_root = build_multi_regression_tree(X_train, auto_attribute_names, auto_discrete_attribute_names)
        
        a_k_err = []
        
        # alpha = 0 --> k_root
        a_rss = evaluate_multi_rss(X_test, k_root)
        a_k_err.append(a_rss)
        
        for alpha in auto_alphas[1:]:
            a_cost, a_node, a_t_rss, a_lvs = find_alpha_multi_pruning_node(X_train, k_root, alpha)
            a_tree = prune_multi_node_subtree(X_train, k_root, a_node)
            
            a_rss = evaluate_multi_rss(X_test, a_tree)
            a_rss /= len(X_test)
            
            a_k_err.append(math.sqrt(a_rss))
            
#             if (k == 0):
#                 print(a_node.label)
        
        for i in range(len(auto_alphas)):
            a_err[i] += a_k_err[i]

    k = K - 1
    l = k * flen

    X_train = X[:l]
    X_test = X[l:]

    k_root = build_multi_regression_tree(X_train, auto_attribute_names, auto_discrete_attribute_names)

    a_k_err = []
    
    # alpha = 0 --> k_root
    a_rss = evaluate_multi_rss(X_test, k_root)
    a_k_err.append(a_rss)

    for alpha in auto_alphas[1:]:
        a_cost, a_node, _, _ = find_alpha_multi_pruning_node(X_train, k_root, alpha)
        a_tree = prune_multi_node_subtree(X_train, k_root, a_node)

        a_rss = evaluate_multi_rss(X_test, a_tree)
        a_rss /= len(X_test)
        a_k_err.append(math.sqrt(a_rss))

    for i in range(len(auto_alphas)):
        a_err[i] += a_k_err[i]
    
    min_err = math.inf
    min_pos = None

    pos = 0
    for err in a_err:
        if (err < min_err):
            min_err = err
            min_pos = pos
        
        pos += 1
    
    print(min_err, min_pos)
    return min_err, min_pos
```
